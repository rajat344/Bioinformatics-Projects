```python
from Bio import Entrez, SeqIO
from Bio.Seq import Seq

# NCBI requires your email (so they can contact if overuse)
Entrez.email = "your@email.com"  # put any real email here

print("NCBI Connection Ready!")
```

```python
print("Fetching real BRCA1 gene from NCBI database...")
print("This is LIVE data from the world's biggest gene database!")
print("=" * 60)

# Fetch it
handle = Entrez.efetch(
    db="nucleotide",
    id="NM_007294",      # BRCA1's official NCBI ID
    rettype="fasta",
    retmode="text"
)

record = SeqIO.read(handle, "fasta")
handle.close()

# Basic info
print("Gene ID     :", record.id)
print("Description :", record.description[:60], "...")
print("DNA Length  :", len(record.seq), "base pairs")
print()

# First 60 letters of DNA
print("DNA (first 60 bases):")
print(record.seq[:60])
print()

# Transcribe
rna = record.seq.transcribe()
print("RNA (first 60 bases):")
print(rna[:60])
print()

# Translate
protein = record.seq.translate()
print("Protein (first 30 amino acids):")
print(protein[:30])
print()

print("=" * 60)
print(f"DONE! Processed real BRCA1 gene with {len(record.seq)} base pairs!")
```

```python
from Bio.SeqUtils import gc_fraction

gc = gc_fraction(record.seq) * 100

print("=" * 60)
print("GC CONTENT ANALYSIS")
print("=" * 60)
print(f"Gene        : BRCA1")
print(f"GC Content  : {gc:.2f}%")
print()

if gc > 60:
    print("Stability: VERY HIGH 🔥 - Hard to denature")
elif gc > 40:
    print("Stability: NORMAL ✅ - Typical human gene")
else:
    print("Stability: LOW ❄️ - Easy to denature")

print()
print("WHY THIS MATTERS:")
print("High GC = more stable DNA = harder for mutations to occur")
print("Used in PCR design, species identification, primer design")
```

```python
print("=" * 60)
print("FINDING START CODON (ATG)")
print("=" * 60)

dna_str = str(record.seq)

# Find where protein coding starts
start_pos = dna_str.find("ATG")
print(f"First ATG (Start Codon) found at position: {start_pos}")
print(f"Sequence around it: ...{dna_str[start_pos-3:start_pos+10]}...")
print()

# Count each base
a = dna_str.count('A')
t = dna_str.count('T')
g = dna_str.count('G')
c = dna_str.count('C')

print("BASE COMPOSITION:")
print(f"  A (Adenine)  : {a} bases ({a/len(dna_str)*100:.1f}%)")
print(f"  T (Thymine)  : {t} bases ({t/len(dna_str)*100:.1f}%)")
print(f"  G (Guanine)  : {g} bases ({g/len(dna_str)*100:.1f}%)")
print(f"  C (Cytosine) : {c} bases ({c/len(dna_str)*100:.1f}%)")
```

```python
# Save everything to a proper report file
with open("BRCA1_analysis_report.txt", "w") as f:
    f.write("BRCA1 GENE ANALYSIS REPORT\n")
    f.write("=" * 60 + "\n")
    f.write(f"Gene ID      : {record.id}\n")
    f.write(f"Description  : {record.description}\n")
    f.write(f"DNA Length   : {len(record.seq)} base pairs\n")
    f.write(f"GC Content   : {gc:.2f}%\n\n")
    f.write(f"DNA Sequence (first 100):\n{record.seq[:100]}\n\n")
    f.write(f"RNA Sequence (first 100):\n{rna[:100]}\n\n")
    f.write(f"Protein Sequence (first 50):\n{protein[:50]}\n")

print("=" * 60)
print("Report saved as BRCA1_analysis_report.txt!")
print("Check your Jupyter folder - file is right there!")
print("=" * 60)
print()
```
