```python
from Bio.Blast import NCBIWWW, NCBIXML
from Bio.Seq import Seq

print("BLAST modules loaded!")
print("Connecting to NCBI BLAST server...")
```

```python
# This is a REAL mystery sequence
# We genuinely don't know what organism this is from
# BLAST will tell us!

mystery_dna = Seq("ATGAAAGCAATTTTCGTACTGAAAGGTTTTGTTGGTTTTCTTGGATCC")

print("Mystery DNA sequence:")
print(mystery_dna)
print(f"Length: {len(mystery_dna)} base pairs")
print()
print("Sending to NCBI BLAST server...")
print("⚠️ This takes 30-60 seconds — BLAST is searching millions of sequences!")
```

```python
# Run BLASTN — DNA vs DNA database
result_handle = NCBIWWW.qblast(
    "blastn",           # type of BLAST
    "nt",               # database (nt = all nucleotides)
    mystery_dna         # our mystery sequence
)

print("BLAST search complete!")
print("Reading results...")

# Parse results
blast_records = NCBIXML.parse(result_handle)
blast_record = next(blast_records)

print(f"Total matches found: {len(blast_record.alignments)}")
```

```python
print("=" * 60)
print("TOP 5 MATCHES FOR OUR MYSTERY DNA")
print("=" * 60)

for i, alignment in enumerate(blast_record.alignments[:5]):
    hsp = alignment.hsps[0]
    
    print(f"\n#{i+1} MATCH:")
    print(f"Organism  : {alignment.title[:60]}...")
    print(f"Score     : {hsp.score}")
    print(f"E-value   : {hsp.expect}")
    print(f"Identity  : {hsp.identities}/{hsp.align_length} bases matched")
    match_percent = (hsp.identities/hsp.align_length) * 100
    print(f"Match %   : {match_percent:.1f}%")
    
    if match_percent > 95:
        print(f"Verdict   : 🔥 VERY STRONG MATCH!")
    elif match_percent > 80:
        print(f"Verdict   : ✅ GOOD MATCH")
    else:
        print(f"Verdict   : ⚠️ WEAK MATCH")
    print("-" * 40)
```

```python
print("=" * 60)
print("FINAL VERDICT")
print("=" * 60)

# Best match
best = blast_record.alignments[0]
best_hsp = best.hsps[0]
best_percent = (best_hsp.identities/best_hsp.align_length) * 100

print(f"Mystery DNA most likely belongs to:")
print(f"{best.title[:80]}")
print()
print(f"Confidence: {best_percent:.1f}% match")
print(f"E-value   : {best_hsp.expect}")
print()
print("E-value explanation:")
print("0.0  = perfect, definitely this organism 🔥")
print("0.05 = strong match ✅")
print("1.0+ = weak, could be random ⚠️")
print()
print("Mystery SOLVED by Python + BLAST! 🧬🔥")
```

```python
from Bio import SeqIO
from Bio.SeqUtils import gc_fraction

# Create FASTA file
#3
with open("genes.fasta", "w") as file:
    file.write(">Gene1\n")
    file.write("ATGCGTACGATCGATCGATC\n")
    file.write(">Gene2\n")
    file.write("ATGCCCGGGAAATTTCCCGG\n")
    file.write(">Gene3\n")
    file.write("GCGCGCGCTATATATAGCGC\n")

# Read FASTA file and calculate GC content
for record in SeqIO.parse("genes.fasta", "fasta"):
    gc = gc_fraction(record.seq) * 100
    print("Gene Name :", record.id)
    print("Sequence  :", record.seq)
    print("GC Content:", round(gc, 2), "%")
    print()
```
